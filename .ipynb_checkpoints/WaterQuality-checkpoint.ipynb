# Data Preprocessing

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing Dataset

```python
dataset =pd.read_csv('water_data.csv',encoding='unicode_escape')
dataset
```

```python
dataset.drop(['STATION CODE'],axis=1,inplace=True)
dataset
```

```python
dataset["STATE"]= dataset["STATE"].apply(lambda x: "" if x=="NAN" else x)
```

```python
dataset["STATE"]
```

```python
dataset["loc"]= dataset["LOCATIONS"]+dataset["STATE"]
```

```python
dataset["loc"]=dataset["loc"].apply(lambda x: x.split(' ')[-1])
```

```python
dataset["loc"].value_counts()
```

```python
dataset.isnull().any()
```

```python
dataset['STATE'].value_counts()
```

```python
s=dataset["STATE"]
dataset["STATE"]=s.apply(lambda x: x.split(' ')[-1])

dataset
```
